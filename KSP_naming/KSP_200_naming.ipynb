```python
import pandas as pd
qa_200= pd.read_csv('/content/drive/MyDrive/AntHouse/0.Data/enterprise/kospi200_50/new/kospi_200.csv')
qa_200.drop(columns='ticker', inplace=True)
```

```python
def naming_all(df,n): #n은 시작할 번호 0~시작
    cnt=0
    num= df.shape[0]
    print("0은 네이밍 시작, 1은 네이밍 종료, 2는 pass 입니다.")
    print('회사당 2개의 네이밍을 진행해주세요.')
    for i in range(n,df.shape[0]):
        print(df.iloc[i,0],'의 네이밍')
        # yn = input('숫자 : ')
        
        # if yn == '2':  # 다음기업 이동
        #     continue
        # elif yn=='1': # 네이밍 종료
        #     break
        # elif yn=='0': # 네이밍 하기 
        for _ in range(2):
            name = input()
            if name=='.': # 해당 턴 종료
                break
            else:
                ls=list(qa_all.iloc[i][1:])
                name = name+f'({df.iloc[i,0]})'
                ls.insert(0,name)
                df.loc[cnt+num] = ls #맨 끝에 누적으로 추가
                cnt += 1
    return df
```

```python
qa = qa_200.copy()
naming_all(qa,0)
```
