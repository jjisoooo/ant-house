```python
def naming_all(df,n):
    cnt=1
    
    for i in range(df.shape[0]):
        print(df.iloc[i,1])
        yn = input()
        if yn == '2': # 다음 기업으로 이동
            continue
        elif yn=='1':  # 네이밍 종료 
            break
        elif yn=='0': # 네이밍 하기 
            for _ in range(n):
                name = input()
                if name=='.': # 해당 턴 종료
                    break
                else:            
                    temp = pd.DataFrame({'ticker':df.iloc[i,0],'name':f'{df.iloc[i,1]}({str(name)})','N.I(억원)':df.iloc[i,2]},index=[49+cnt])
                    df = pd.concat([df,temp])
                    cnt += 1
    

    return df
```

```python
dn = 다나.copy()
dn = naming_all(dn,0)
```

```python
dn.to_csv('100_150.csv', index=False)
```
